```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Create figure and axis
fig, ax = plt.subplots(figsize=(14, 10))

# Components positions
components = {
    "User Interface (UI)": (0.1, 0.8),
    "Frontend (Client-Side)": (0.5, 0.8),
    "Backend (Server-Side)": (0.9, 0.8),
    "Database": (0.5, 0.4),
    "API Integrations": (0.9, 0.4),
    "Data Analytics": (0.3, 0.1),
    "Security": (0.1, 0.4),
    "Deployment": (0.7, 0.1)
}

# Draw components as rectangles
for comp, (x, y) in components.items():
    rect = mpatches.FancyBboxPatch((x - 0.05, y - 0.05), 0.1, 0.1, boxstyle="round,pad=0.1", edgecolor="black", facecolor="lightblue")
    ax.add_patch(rect)
    ax.text(x, y, comp, ha="center", va="center", fontsize=10)

# Define connections
connections = [
    ("User Interface (UI)", "Frontend (Client-Side)"),
    ("Frontend (Client-Side)", "Backend (Server-Side)"),
    ("Backend (Server-Side)", "Database"),
    ("Backend (Server-Side)", "API Integrations"),
    ("Backend (Server-Side)", "Data Analytics"),
    ("Frontend (Client-Side)", "Security"),
    ("Backend (Server-Side)", "Security"),
    ("Deployment", "Backend (Server-Side)"),
    ("Deployment", "Frontend (Client-Side)")
]

# Draw connections as arrows
for start, end in connections:
    start_x, start_y = components[start]
    end_x, end_y = components[end]
    ax.annotate("", xy=(end_x, end_y), xytext=(start_x, start_y), arrowprops=dict(arrowstyle="->", lw=1.5))

# Set limits and remove axes
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Show plot
plt.show()
```
